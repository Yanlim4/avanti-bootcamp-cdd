# Visualização

## Barra

### Importando Bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Leitura de Dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')

df.head()
```

### Gráfico

```python
#Simplificando valores
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)

#Ordenando as barras
df = df.sort_values(by='arrecadacao', ascending=False)

#Ativando linhas verticais
sns.set_style("whitegrid")

#Padronizando a cor das barras e setando como barras horizontais
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='blue')

#Adicionando valores nas barras
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

#Melhorando o titulo e os labels dos eixos
plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')

#Adicionando uma linha vertical na origem e removendo bordas
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```

## Histograma

### Importando Bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Leitura de Dados

```python
df = sns.load_dataset("penguins")
df.head()
```

### Gráfico

```python
#Inserindo linhas verticais e horizontais
sns.set_style("whitegrid")

#Melhorando o titulo e os labels dos eixos
ax = sns.histplot(data=df, x='body_mass_g', color='green')
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')

# adicionando uma linha a origem e removendo as bordas
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```

## Boxplot

### Importando Bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

### Leitura de Dados

```python
df = sns.load_dataset('tips')
df.head()
```

### Gráficos

```python
#Linhas verticais
sns.set_style("whitegrid")

#Padronizando a cor das barras
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x="day", y="total_bill", data=df, color="orange")

#Melhorando o titulo e os labels do eixo
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
